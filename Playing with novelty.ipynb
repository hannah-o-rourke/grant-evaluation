```python
import pandas as pd

```

```python
url = 'https://raw.githubusercontent.com/hannah-o-rourke/grant-evaluation/main/variables.csv'
```

```python
df1 = pd.read_csv(url)
```

```python
df1.shape
```

```python
df1[['Name','Novelty']]
```

```python
df1['Grant'] = 78947
df1
```

```python
def novelty_groups(x):
    if x < 1:
        return 'less'
    else:
        return 'more'
df1['Novelty_Groups'] = df1['Novelty'].apply(novelty_groups)
print(df1)
```

```python
df1.loc[df1.Novelty == 1, 'Grant'] = "More"
df1.loc[df1.Novelty == 0, 'Grant'] = "Less"
df1
```
